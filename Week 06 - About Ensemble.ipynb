# Week 6 - About Ensemble

## 앙상블 (Ensemble)
'조화'라는 뜻의 불어

- 여러 개의 분류기(Classifier)를 생성하고, 그 예측을 결합함으로써 보다 정확한 예측을 도출하는 기법.

- 앙상블의 유형은 일반적으로 보팅(Voting), 배깅(Bagging), 부스팅(Boosting), 스태킹(Stacking) 등이 있다.

- 대표적인 배깅(Bagging)은 랜덤포레스트(RandomForest) 알고리즘이 있고, 부스팅(Boosting)은 AdaBoosting, Gradient Boosting, XGBoost, LightGBM 등이 있다.

<b> 특징 </b>

- 단일 모델의 약점을 다수 모델을 결합하여 보완.

- 뛰어난 성능을 가진 모델들로만 구성하는 것보다 성능이 떨어지더라도 서로 다른 유형의 모델을 섞는 것이 오히려 전체 성능이 좋아질 수 있음.

- 랜덤포레스트나 뛰어난 부스팅 알고리즘은 모두 결정 트리(Decision Tree) 알고리즘을 기반으로 형성됨.

- 결정 트리(Decision Tree)의 단점인 Overfitting을 수 백, 수 천개의 많을 분류기를 결합해 보완하여 장점인 직관적인 분류 기준은 강화된다.

## Voting vs Bagging

이미지 출처 : https://velog.io/@jiselectric/Ensemble-Learning-Voting-and-Bagging-at6219ae

보팅(Voting)과 배깅(Bagging) 모두 분류기를 결합하여 최종 예측을 하는 것은 동일하다.

- 보팅(Voting)은 동일한 데이터셋에서 서로 다른 분류기를 통해 학습, 예측한다.
- 배깅(Bagging)은 기존 데이터에서 샘플링을 통해 얻어진 데이터셋을 동일 분류기들을 통해 학습, 예측
