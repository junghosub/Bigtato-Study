# Week 5 - Metrics of Classification

모델을 만들면 해당 모델이 좋은 모델인지 평가해야한다. 

분류 모델의 경우, 대표적으로 사용되는 평가지표는 Confusion Matrix와 ROC curve, AUC가 있다.

Confusion Matrix, ROC Curve, AUC에 대한 개념을 알아보고, Week 4의 Pima Indian 예제를 통해 확인

### Confusion Matrix (혼동행렬)

#### 정확도 (Accuracy)

전체 예측 건수에서 정답(TP, TN)을 맞힌 건수의 비율이다.

#### 재현율(Recall)

실제로 정답이 True 인 것들 중에서 분류기가 True로 예측한 비율이다. 그래서 애초에 True가 발생하는 확률이 적을 때 사용하면 좋다.

#### 정밀도 (Precision)
True라고 예측한 것들 중에서 실제로 True인 비율이다.

#### 특이도 (Specificity)
실제 False인 것들 중 분류기가 False라고 예측한 비율

#### F1 Score
재현율(Recall)과 정밀도(Precision)는 반비례적 관계이다.(Trade off)

이를 보완하기 위해 나온 것이 F1 Score이다.

### ROC Curve and AUC

#### ROC curve

y축은 재현율(Recall), x축은 1 - 특이도(Specificity)로 그래프를 그린 것.

image link : https://glassboxmedicine.files.wordpress.com/2019/02/roc-curve-v2.png?w=576

ROC curve에 대해 잘 설명한 글 : https://angeloyeo.github.io/2020/08/05/ROC.html

####  AUC(area under the curve)
ROC curve의 넓이를 구한 것이다. 0.5 <= AUC <= 1로 나타나며, 1에 가까울수록 좋은 분류모델이다.
